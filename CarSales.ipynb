```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

brands = ["Tesla", "BYD", "BMW", "Volkswagen", "Hyundai", "Kia", "Ford", "Mercedes", "Nissan", "Volvo"]
sales_2023 = [1800, 1500, 700, 850, 700, 650, 600, 580, 500, 480]
sales_2024 = [2000, 1750, 950, 900, 780, 700, 670, 600, 520, 510]


x = np.arange(len(brands))
width = 0.4


plt.bar(x - width/2, sales_2023, width, label='2023', color='skyblue')
plt.bar(x + width/2, sales_2024, width, label='2024', color='orange')


plt.xticks(x, brands, rotation=45, ha='right')
plt.ylabel("EV Sales (in thousands)")
plt.title("Top Electric Car Sales by Brand (2023 vs 2024)")
plt.legend()
plt.tight_layout()
plt.show()
```
